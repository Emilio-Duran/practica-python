### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

ventas_totales = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

ventas_totales['Fecha de Compra'] = pd.to_datetime(ventas_totales['Fecha de Compra'], dayfirst=True)
```

```python
facturacion_total = ventas_totales['Precio'].sum()
print(f"La facturación total de todas las tiendas es: ${facturacion_total:,.2f}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria = ventas_totales.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría:")
print(ventas_por_categoria)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = ventas_totales['Calificación'].mean()
print(f"La calificación promedio global es: {calificacion_promedio:.2f} estrellas")
```

# 4. Productos más y menos vendidos

```python
# El más vendido
top_producto = ventas_totales['Producto'].value_counts().idxmax()
# El menos vendido
bottom_producto = ventas_totales['Producto'].value_counts().idxmin()

print(f"El producto estrella es: {top_producto}")
print(f"El producto con menos salida es: {bottom_producto}")
```

# 5. Envío promedio por tienda

```python
tiendas_dict = {
    'Tienda 1': tienda['Costo de envío'].mean(),
    'Tienda 2': tienda2['Costo de envío'].mean(),
    'Tienda 3': tienda3['Costo de envío'].mean(),
    'Tienda 4': tienda4['Costo de envío'].mean()
}

envio_promedio = pd.Series(tiendas_dict)
print("Costo de envío promedio por cada tienda:")
print(envio_promedio)
```
